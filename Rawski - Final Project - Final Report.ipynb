# Introduction & Motivation
For this project, I analyzed my personal Amazon Music listening data to explore patterns in how I engage with music throughout the day. While initially interested in studying public transportation data, limitations in data availability led me to shift focus toward a dataset I interact with daily: my music listening history. This shift allowed me to ask a question rooted in my own habits and routines, particularly surrounding when and how I listen to music. The driving question guiding this analysis was: **How does the time of day that I listen to music affect how much of a song I actually listen to?** <br>
<br>
To answer this question, I downloaded my personal listening history directly from Amazon Music. After submitting a data request, I received a CSV file containing detailed records of my listening activity. With this, I was especially interested in comparing listening behavior during my daily commute versus nighttime listening.<br>
<br>
My initial hypothesis was that I would finish fewer songs during my commute, due to distractions and time constraints, and finish more songs at night when listening more passively or as background noise.

# Data Structure & Cleaning

The dataset provided by Amazon Music is a CSV file containing 8,928 rows and 12 columns, covering my listening activity from October 13, 2024 to November 11, 2025. Each row represents a single playback event. The data includes a mix of timestamps, categorical text data, and numerical values. Key numerical variables included track length and listening duration, while categorical variables included playback stop reason and listening period.

Before analysis, I cleaned the dataset by removing columns that were not relevant to answering my research question, such as device and territory identifiers. I then transformed the timestamp data to create new columns representing the hour of playback, general time of day, and broader listening periods (commuting hours, nighttime, or other).

Finally, I calculated a completion percentage for each song by comparing how long I listened to the track with its total length. These changes, visible in Figure 1, allowed me to focus the dataset on time-based listening behavior and song completion, which were central to my analysis.

# Analysis

The goal of this analysis was to determine whether the time of day I listened to music affected how much of each song I completed, with a specific focus on comparing listening during my daily commute versus nighttime listening.<br>
<br>
On average the initial statistics showed, the songs in my dataset were just over three minutes long, and I typically listened to each song for nearly that full duration. This suggested that, overall, I tend to finish most songs I start. This observation was further supported by the Reason for Stopped Playback data, which showed that approximately 87% of all playback events ended because the track finished naturally rather than being skipped or stopped early.<br>
<br>
The most important findings emerged when comparing completion behavior across listening periods. Completion rate was used as the primary metric because it directly reflects how fully I engage with a song during different times of day. By calculating the average completion rate for songs played during commuting hours, nighttime listening, and other times of day, clear differences became visible. Contrary to my expectations, songs listened to during my commute had a slightly higher average completion rate than those listened to at night. This pattern is clearly shown in Figure 5, which compares average completion rates by listening period.

# Conclusion

The results of this analysis do not support my original hypothesis. I expected that listening to music at night would result in higher song completion rates than listening during my commute. However, the data shows the opposite trend: songs played during commuting hours were completed at a slightly higher rate than those played at night.<br>
<br>
This finding highlights the value of data-driven analysis, as my assumptions about my own behavior did not match what the data revealed. While nighttime listening felt more intentional, the structured nature of commuting may actually encourage more consistent, uninterrupted listening. Overall, this project demonstrates how personal data can uncover patterns that are not immediately obvious through observation alone.

# Limitations & Future Decisions

While this analysis provided meaningful insights, there are several limitations to consider. First, the data represents only one individual, my listening behavior, which limits how broadly the results can be generalized. Additionally, listening periods were defined using fixed time ranges, which may not perfectly capture the true context of each listening session. For example, not all listening during commuting hours necessarily occurred while actively commuting. I think that if I could have further filtered that data to reflect my schedule, categorizing my hours could have been more accurate.<br>
<br>
There were also some inconsistencies in the data, such as playback events where the recorded listening duration exceeded the actual track length. These anomalies likely result from background playback or system recording issues and may slightly affect completion calculations. Further refinement of this category by defining a rule that the listen duration cannot exceed the track duration, may help for prevention of this type of error in the future.<br>
<br>
Future analyses could improve on this project by incorporating additional contextual data, such as location, activity type, or playlist usage. Comparing multiple users data or extending the analysis to include genre or mood-based listening could also provide deeper insights into how listening behavior changes across different situations. Other expansions could include comparing listening data across multiple music platforms, such as Spotify or Apple Music. Comparing platforms could help determine whether situational listening behavior is influenced by platform design, user interface, or recommendation systems, rather than time of day alone. This would provide a broader understanding of how different platforms shape user interaction with music in everyday contexts.
