# Section 3 Data in a Pandas DataFrame

```python
# imports
import requests
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'/Users/morawski28/Downloads/Amazon-Music/listening.csv')
df.head()
```

```python
df = df.drop(columns=["ASIN", "Device Type", "Music Territory", "Title Set ASIN", "Artist ASIN", "Selection Source Artist ASIN", "Playback from Shuffle"])
df.head(30)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

def time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df['Time of Day'] = df['Date'].dt.hour.map(time_of_day)

df["Time of Day"].value_counts()

df.head(30)
```

```python
cols = df.columns.tolist()

# Remove column if it already exists in the list (prevents duplicates)
cols.remove("Time of Day")

# Find where "date" is located
Date_index = cols.index("Date")

# Insert time_of_day right after date
cols.insert(Date_index + 1, "Time of Day")

# Reorder the dataframe
df = df[cols]

df.head(30)
```

# Section 5 Analysis

My plan for my analysis is see if there is a correlation between the Time of Day/Date column and the Reason for Stopped Playback column to determine if I do really listen to more songs fully at night due to a lack of interruptions like there are during my daily commute. What I plan to do is change the criteria of my Time of Day labels to reflect the time of my commute, then based on these time labels check to see what time I had the most TrackFinished labels in the Reason for Stopped Playback. 

```python
# Adding/Changing my columns so I can determine the hour, time of day, and listening period the data was collected at
df["hour"] = df["Date"].dt.hour

df["listening_period"] = df["hour"].apply(
    lambda h: "Commute" if (6 <= h < 11) or (15 <= h < 18)
    else ("Night" if 19 <= h < 24 else "Other")
)

df.head(30)
```

```python
cols = df.columns.tolist()

# Remove column if it already exists in the list (prevents duplicates)
cols.remove("hour")
cols.remove("listening_period")

# Find where "Time of Day" is located
Date_index = cols.index("Time of Day")

# Insert hour right after date
cols.insert(Date_index + 1, "hour")
cols.insert(Date_index + 2, "listening_period")

cols.remove

# Reorder the dataframe
df = df[cols]

df.head(30)
```

```python
# Determines if each row (song) was played all of the way through and what percentage it was played through
percent_played = ((df["Listen Duration in Milliseconds"] / df["Track Length in Milliseconds"])*100)
percent_played
```

```python
df["percent_played"] = df["Listen Duration in Milliseconds"] / df["Track Length in Milliseconds"]
df.head(30)
```

```python
df["percent_played_formatted"] = (df["percent_played"] * 100).round(2).astype(str) + "%"
formatted = df["percent_played_formatted"]
df = df.drop(columns=["percent_played_formatted"])
df["percent_played_formatted"] = formatted
df.head(30)
```

```python
# Sorting, Calculating and Determining the completion rates of songs based on the assignment for the Reason for Stopped Playback sorted bsed on the listening period I established. 
completion_rates = (
    df.groupby("listening_period")["Reason for Stopped Playback"]
      .apply(lambda s: (s == "trackFinished").mean() * 100)
)
completion_rates
```

```python
df["Track Length in Milliseconds"] = df["Track Length in Milliseconds"] / 60000
df["Track Length in Milliseconds"].hist()
plt.xlabel("Track Length (minutes)")
plt.ylabel("Number of Tracks")
plt.title("Distribution of Track Lengths")
plt.show()
```

```python
df["Listen Duration in Milliseconds"] = df["Listen Duration in Milliseconds"] / 60000
df["Listen Duration in Milliseconds"].hist()
plt.xlabel("Listen Duration (minutes)")
plt.ylabel("Number of Tracks")
plt.xlim(0, 10)
plt.title("Distribution of Listen Duration")
plt.show()
```

```python
reason_counts.plot(kind="bar")
plt.xlabel("Reason for Stopped Playback")
plt.ylabel("Count")
plt.title("Playback Stop Reasons")
plt.show()
```

```python
reason_percent.plot(kind="bar")
plt.xlabel("Reason for Stopped Playback")
plt.ylabel("Percent")
plt.title("Playback Stop Reasons")
plt.show()
```

```python
# Group by listening period and compute mean percent completion
completion_rates = df.groupby("listening_period")["percent_played"].mean()

completion_rates.plot(kind="bar")
plt.ylabel("Average Completion Rate")
plt.xlabel("Listening Period")
plt.title("Completion Rate by Listening Period")
plt.ylim(0, 1)  # because percent_played is between 0–1
plt.show()
```
